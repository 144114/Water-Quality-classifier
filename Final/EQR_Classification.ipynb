```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Waterbase_v2018_1_WISE4_csv/Waterbase_v2018_1_T_WISE4_BiologyEQRClassificationProcedure.csv") # Read File
```

```python
# Divide the citeria into 2 categories 

data_LW = data[data["parameterWaterBodyCategory"] == "LW"]  # LW 
data_RW = data[data["parameterWaterBodyCategory"] == "RW"]  # RW
```

```python
# Group the country code together

data_LW = data_LW.groupby(["CountryCode"]).mean()
data_RW = data_RW.groupby(["CountryCode"]).mean()
```

```python
# Check the Value in each Dataset

data_LW
```

```python
# Check the Value in each Dataset

data_RW
```

```python
# The value that we are going to keep from this dataset are these following column

# 1. parameterICStatusOfDeterminandBiologyEQR (Flag indicating whether the biological determinand has been intercalibrated)
# 2. parameterBoundaryValueClasses12 (Boundary between high and good EQR)
# 3. parameterBoundaryValueClasses23 (Boundary between good and moderate EQR)
# 4. parameterBoundaryValueClasses34 (Boundary between moderate and poor EQR)
# 5. parameterBoundaryValueClasses45 (Boundary between poor and bad EQR)

data_LW = data_LW[["parameterICStatusOfDeterminandBiologyEQR","parameterBoundaryValueClasses12","parameterBoundaryValueClasses23","parameterBoundaryValueClasses34","parameterBoundaryValueClasses45"]]
data_RW = data_RW[["parameterICStatusOfDeterminandBiologyEQR","parameterBoundaryValueClasses12","parameterBoundaryValueClasses23","parameterBoundaryValueClasses34","parameterBoundaryValueClasses45"]]
```

```python
# Reset Indexes
data_LW = data_LW.reset_index()
data_RW = data_RW.reset_index()
```

```python
# Fill the missing Value
# First, Focusing on data_LW dataset
data_LW
```

```python
# BE and CY rows has a missing parameterBoundayValueClasses12 but the overall data look really similar to AT, FI and SI. Therefore, in this case the value will be filled by using parameterBoundayValueClasses12 of other column
data_LW.iloc[[0,1,3,6,17],1:]
```

```python
# BE and CY rows has a missing parameterBoundayValueClasses12 but the overall data look really similar to AT, FI and SI. Therefore, in this case the value will be filled by using parameterBoundayValueClasses12 of other column
data_LW.iloc[1,1:] = data_LW.iloc[0,1:] # Set BE = AT
data_LW.iloc[3,1:] = data_LW.iloc[0,1:] # Set BE = AT
```

```python
data_LW
```

```python
# By checking the EQR_clean dataset we gain an information that there is no parameterWaterBodyCategory in ES and PT country that contain LW type of water on it therefore, we can drop these rows
data_LW = data_LW.drop(labels=[5,14], axis=0)
```

```python
data_LW
```

```python
# Second, Focusing on data_RW dataset
data_RW
```

```python
# For this dataset the only rows that contain is CH but we have no data that contain CH country code on the EQR_clean dataset Therefore we can drop this dataset

data_RW = data_RW.drop(labels=3, axis=0)
```
